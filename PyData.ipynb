```python
#Import H2O Package 
import h2o
from h2o.estimators.gbm import H2OGradientBoostingEstimator
```

```python
#Initialize instance of H2O 
h2o.init(ip = "172.20.10.3", port = 54321, strict_version_check= False)
```

```python
#Set file path
path = "/home/test/pydata-data/loganalysisdata.txt"
```

```python
#Import the data
data = h2o.import_file(path=path)
```

```python
data.columns = ["priority_id", "date_format", "string_length", "num_ip_addresses", "num_mac_addresses", "log_class", 
"four_letter_word_ratio", "five_letter_word_ratio", "six_letter_word_ratio", "seven_letter_word_ratio",
"eight_letter_word_ratio", "nine_letter_word_ratio", "ten_letter_word_ratio", "eleven_letter_word_ratio",
"twelve_letter_word_ratio", "thirteen_letter_word_ratio", "fourteen_letter_word_ratio", "fifteen_letter_word_ratio"]
```

```python
data.col_names
```

```python
#Do a test-train splite (80-20 split)
s = data["priority_id"].runif()
train = data[s <= 0.80]
test  = data[s > 0.80]
```

```python
#Define your response and predictor variables
x = ["priority_id", "date_format", "string_length", "num_ip_addresses", "num_mac_addresses", 
"four_letter_word_ratio", "five_letter_word_ratio", "six_letter_word_ratio", "seven_letter_word_ratio",
"eight_letter_word_ratio", "nine_letter_word_ratio", "ten_letter_word_ratio", "eleven_letter_word_ratio",
"twelve_letter_word_ratio", "thirteen_letter_word_ratio", "fourteen_letter_word_ratio", "fifteen_letter_word_ratio"]
y = "log_class"
```

```python
#Set parameters for GBM model
model = H2OGradientBoostingEstimator(model_id="DrTsModel",
                                       score_each_iteration= False,
                                       ntrees=50,
                                       learn_rate=0.05, 
                                        nfolds = 5)
```

```python
#Build the GBM model
model.train(x=x, y=y, training_frame=train, validation_frame=test)
```

```python
#View model results
model
```

```python
#Look at the variable importances
model.varimp()
```

```python
#Generate confusion matrix for test data.
mod_out = model.confusion_matrix(test)
```
